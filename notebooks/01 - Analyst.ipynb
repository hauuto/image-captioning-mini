### 1. Load dataset

```python
import os
import pandas as pd
```

```python
def check_flickr8k(root_dir):
    captions_path = os.path.join(root_dir, "captions.txt")
    images_dir = os.path.join(root_dir, "Images")
    if not os.path.isdir(images_dir):
        images_dir = os.path.join(root_dir, "images")

    df = pd.read_csv(captions_path)

    df["image"] = df["image"].str.strip()
    df["caption"] = df["caption"].str.strip()

    print("Rows:", len(df))
    print("Unique images:", df["image"].nunique())

    empty_captions = (df["caption"] == "").sum()
    print("Empty captions:", int(empty_captions))

    missing = []
    for name in df["image"].unique():
        path = os.path.join(images_dir, os.path.basename(name))
        if not os.path.exists(path):
            missing.append(name)

    print("Missing image files:", len(missing))
    for m in missing[:20]:
        print(" ", m)

    counts = df.groupby("image")["caption"].count()
    print("Captions per image:")
    print(" min =", int(counts.min()))
    print(" max =", int(counts.max()))
    print(" mean =", float(counts.mean()))
```

```python
check_flickr8k(flickr_path)
```
