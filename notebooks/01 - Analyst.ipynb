## Flickr 8k

### Download dataset

```python
import kagglehub
import shutil
import os

cache_path = kagglehub.dataset_download("adityajn105/flickr8k")

print(f"Dataset downloaded to cache at: {cache_path}")

target_path = "../data/flickr8k"

if os.path.exists(target_path):
    print(f"Directory {target_path} already exists. Please remove it or choose a new path.")
else:
    shutil.copytree(cache_path, target_path)
    print(f"Files moved to: {target_path}")
```
